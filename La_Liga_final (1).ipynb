```python
import matplotlib.pyplot as plt
from matplotlib import pyplot
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("LaLiga_dataset.csv")
```

```python
pd.DataFrame(data)
```

```python
#getting the data for season 2016-17
season2017 = data[data['season']== '2016-17']


pd.DataFrame(season2017)
```

```python
#getting just the points for season 2016-17
points = np.array(season2017['points'])
print(points)

#making the size of the graph bigger 
plt.figure(figsize=(10,10))

#type of the graph line
plt.plot(points, c="red", ls="--", marker = "s", ms= 3)

#graph name
plt.title('Season 2017 Teams Points')

#showing the graph
plt.show()
```

```python
#getting the team name and the points for season 2016-17
po_team = np.array([season2017['club'],season2017['points']])
```

```python
#storing the team names as X
x = po_team[0]
#storing the team pints as Y
y=po_team[1]

#graph size
plt.figure(figsize=(10,10))
#graph title
pyplot.title('Points for season 2016-17')
#graph type
pyplot.bar(x,y)
#graph x value rotation
plt.xticks(rotation=90)
```

```python
#graph size
plt.figure(figsize=(10,10))
#graph title
pyplot.title('Points for season 2016-17')
#graph type
pyplot.scatter(y,x)
```

```python
#getting the highlights from season 2016-17 for Barcelona team
barcelona= season2017[season2017['club'] == 'Barcelona']
pd.DataFrame(barcelona)
```

```python
#storing the goals scored at games played home for Barcelona
bar_home_goal = int(barcelona['home_goals'])
#storing the goals scored at games played away for Barcelona
bar_away_goal = int(barcelona['away_goals'])
#storing the goals for the whole season 2016-17 for Barcelona
bar_total_goal = int(barcelona['goals_scored'])

#storing the X with the goals type
x = ['Home', 'Away', 'Total']
#storing the Y with the goals
y = [int(bar_home_goal), int(bar_away_goal), bar_total_goal]
# print(y)
# print(x)



# graph size
plt.figure(figsize=(10,10))
#graph title
pyplot.title('BARCELONA- goals')
#graph type
plt.plot(x, y, label='New Line!')
```

```python
#different type of graph to show the goals

#graph size
plt.figure(figsize=(10,10))
#graph title
pyplot.title('BARCELONA- goals')
#graph type
pyplot.scatter(x,y)
```

```python
#different type of graph to show the goals

#graph size
plt.figure(figsize=(10,10))
#graph title
pyplot.title('BARCELONA- goals')
#graph type
pyplot.bar(x,y)
```

```python
#getting the total of goals scored for the season 2016-17 for each team
goal_team = np.array([season2017['club'],season2017['goals_scored']])

#storing the team names at X
x = goal_team[0]
#stoting the team goals at Y
y=goal_team[1]

#graph size
plt.figure(figsize=(10,10))
#graph title
pyplot.title('Team goals scored for season 2016-17')
#graph type
pyplot.bar(x,y)
#rotating the x value
plt.xticks(rotation=90)
```

```python
#getting the dat for all the teams for season 2015-16
season2016 = data[data['season']== '2015-16']

pd.DataFrame(season2016)
```

```python
#showing the data for home and away goals for season 2015-16 with seaborn graph-plot
sns.lmplot("home_goals", "away_goals", season2016,height=10, aspect=0.6,scatter_kws ={"marker":"o","color":"b"}
          ,line_kws={"linewidth":1,"color":"deeppink"})
#setting title for the plot
ax = plt.gca()
ax.set_title("Season 2015-2016")
```

```python
#showing the data for home and away goals for season 2016-17 with seaborn graph-plot
sns.lmplot("home_goals", "away_goals", season2017,height=10, aspect=0.6,scatter_kws ={"marker":"o","color":"black"}
          ,line_kws={"linewidth":1,"color":"r"})
#setting title for the plot
ax = plt.gca()
ax.set_title("Season 2016-2017")
```

```python
#showing the data for home and away goals for season 2015-16 with seaborn graph-plot
sns.lmplot("home_goals","away_goals",season2016,order=4, scatter_kws ={"marker":"o","color":"c"}
          ,line_kws={"linewidth":1,"color":"black"}, height=10, aspect=0.6);

#setting title for the plot
ax = plt.gca()
ax.set_title("Season 2015-2016")
```

```python
#showing the data for home and away goals for season 2016-17 with seaborn graph-plot
sns.lmplot("home_goals","away_goals",season2017,order=4, scatter_kws ={"marker":"o","color":"c"}
          ,line_kws={"linewidth":1,"color":"black"}, height=10, aspect=0.6);
#setting title for the plot
ax = plt.gca()
ax.set_title("Season 2016-2017")
```

```python
#getting all the teams and points for all the seasons from 1970-71 to 2016-17
team_dframe = data.pivot("club","season","points")
team_dframe
```

```python
#showing all the teams and points for all the seasons from 1970-71 to 2016-17

#size
f, ax = plt.subplots(figsize=(40, 28))
#showing the data in a heatmap
sns.heatmap(team_dframe,annot=True)
```

```python
#showing all the teams and points for all the seasons from 1970-71 to 2016-17

#size
f, ax = plt.subplots(figsize=(30, 21))
#showing the data in a heatmap
sns.heatmap(team_dframe,center=team_dframe.loc['Barcelona','2012-13'],annot=True)
```

```python
#showing all the teams and points for all the seasons from 1970-71 to 2016-17
f, ax = plt.subplots(figsize=(30, 21))

cmap = sns.cubehelix_palette(rot=-.8, as_cmap=True)
ax = sns.scatterplot(x="season", y="club",
                     hue="home_goals", size="away_goals",
                     palette=cmap, sizes=(30, 400),
                     data=data)
plt.xticks(rotation=45)
```
